```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from functools import reduce
```

```python
def f(k,x):
    return np.cos(k*x)
```

```python
x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, f(1, x))
plt.plot(x, f(0.5, x))
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```python
a = np.array([10,10,10])
b = np.array([0,0,-10])
```

```python
def vector_length(v):
    r_2 = reduce(lambda a, x: a + x**2, v, 0)
    return np.sqrt(r_2)
```

```python
vector_length(a)
```

```python
vector_length(b)
```

```python
def circle(x):
    return np.sqrt(1-x**2)
```

```python
x = np.linspace(-1.0, 1.0, 100)
plt.plot(x, circle(x))
plt.plot(x, -circle(x))
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```
